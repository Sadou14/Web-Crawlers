```python
web = "http://books.toscrape.com/?page=%s" 
for page in range(1,8):
    url = web % page
    data = html.fromstring(requests.get(url).text)
# collect Full title,More information URL ,Price, Cover URL, Rating (number of stars)
for review in data.xpath("//li[@class='col-xs-6 col-sm-4 col-md-3 col-lg-3']"):
    title = review.xpath('article/h3/*/@title')
    info = review.xpath('article/h3/*/@href')
    price = review.xpath('article/div/p[1]/text()')
    cover_url = review.xpath("article/div[@class='image_container']/a/@href")
    rating= review.xpath('article/p/@class')    
    print(title, info, price, cover_url, rating)
```

```python
finaldata=[]

web = "http://books.toscrape.com/?page=%s" 

for page in range(1,8):
    url = web % page
    data = html.fromstring(requests.get(url).text)
    
    # collect Full title,More information URL ,Price, Cover URL, Rating (number of stars)
    for review in data.xpath("//li[@class='col-xs-6 col-sm-4 col-md-3 col-lg-3']"):
        title = review.xpath('article/h3/*/@title')
        #info = review.xpath('article/h3/*/@href')
        price = review.xpath('article/div/p[1]/text()')
        cover_url = review.xpath("article/div[@class='image_container']/a/@href")
        rating= review.xpath('article/p/@class')
        print(title, price, cover_url, rating)
        
        finaldata.append([title, price, cover_url, rating])
        
len(finaldata)
```
